```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# setting styles
plt.style.use('classic')
```

```python
import numpy as np
```

```python
# 0 to 10 tak 100 points plot krne ha
x = np.linspace(0,10,100)

fig = plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')

```

```python
# saving the fig in a specific format
fig.savefig('myfig.jpg')
```

```python
# soft figures

# create first panel
plt.figure()
plt.subplot(2,1,1) #(row,cols,panel number)
plt.plot(x,np.sin(x))

# create second panel
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
#  grid
# ek hi figure mein dono plot chahiye

fig,ax = plt.subplots(2) # for grid

ax[0].plot(x,np.sin(x))

ax[1].plot(x,np.cos(x))

```
